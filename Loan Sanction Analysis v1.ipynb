```python
#18. Check outliers in Loan amount and credit history column and replace them with mean.
#identify the mean
LoanAmount_mean = df['LoanAmount'].mean()
Credit_History_mean =df['Credit_History'].mean()

#identify the SD
LoanAmount_std = df['LoanAmount'].std()
Credit_History_std =df['Credit_History'].std()

#Assuming outlier threshold
outlier_threshold = 3

loanAmount_outliers = (df['LoanAmount']-LoanAmount_mean).abs()>outlier_threshold*LoanAmount_std
Credit_History_Outliers =(df['Credit_History']-Credit_History_mean).abs()>outlier_threshold*Credit_History_std

#Replacing outliers with the mean value
df.loc[loanAmount_outliers,'LoanAmount']=LoanAmount_mean
df.loc[Credit_History_Outliers,'Credit_History']=Credit_History_mean
```
